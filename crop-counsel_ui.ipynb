```python
import pyttsx3                                                            
import pandas as pd                                                       
from sklearn import preprocessing                                         
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier                        
import numpy as np                                                        
import PySimpleGUI as sg 
from sklearn.svm import SVC
```

```python
df = pd.read_csv(r"C:/Users/sunny/OneDrive/Desktop/dataset/crops1.csv", header = 0)                            
print(df)                                                              
print(df.shape)  
```

```python
engine = pyttsx3.init('sapi5')                                          
voices = engine.getProperty('voices')
rate = engine.getProperty('rate')
engine.setProperty('rate', rate-20)
engine.setProperty('voice',voices[0].id)


def speak(audio):
    engine.say(audio)
    engine.runAndWait()

le = preprocessing.LabelEncoder()                                         
crop = le.fit_transform(list(df["label"]))                              

                                                    
NITROGEN = list(df["N"])                                         
PHOSPHORUS = list(df["P"])                                    
POTASSIUM = list(df["K"])                                      
TEMPERATURE = list(df["temperature(in degree celsius)"])                                  
HUMIDITY = list(df["humidity(in percentage)"])                                        
PH = list(df["ph of soil"])                                                    
RAINFALL = list(df["rainfall( in mm )"]) 
WATER_AVAILIBILITY = list(df["water-availability(liters per year)"])


features = list(zip(NITROGEN, PHOSPHORUS, POTASSIUM, TEMPERATURE, HUMIDITY, PH, RAINFALL, WATER_AVAILIBILITY))                     
features = np.array([NITROGEN, PHOSPHORUS, POTASSIUM, TEMPERATURE, HUMIDITY, PH, RAINFALL, WATER_AVAILIBILITY])                    

features = features.transpose()                                                                                 
print(features.shape)                                                                                          
print(crop.shape)
                                                    
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_grid={}
model = RandomizedSearchCV(RandomForestClassifier(), param_distributions=param_grid, n_iter=100, cv=5, verbose=2, random_state=42, n_jobs=-1)
model.fit(features, crop)
```

```python
layout = [[sg.Text('                                   CropCounsel', font=("Helvetica", 30), text_color = 'orange')],                                                                                                            
         [sg.Text('Please enter the following details :-', font=("Helvetica", 20))],                                                                                          
         [sg.Text('Enter ratio of Nitrogen in the soil                                  :', font=("Helvetica", 20)), sg.Input(font=("Helvetica",20), size = (20,1) ), sg.Text('0 To 500', font=("Helvetica", 20))],
         [sg.Text('Enter ratio of Phosphorous in the soil                           :', font=("Helvetica", 20)), sg.Input(font=("Helvetica", 20),size = (20,1)), sg.Text('0 To 250', font=("Helvetica", 20))],
         [sg.Text('Enter ratio of Potassium in the soil                               :', font=("Helvetica", 20)), sg.Input(font=("Helvetica", 20),size = (20,1)), sg.Text('0 To 250', font=("Helvetica", 20))],
         [sg.Text('Enter average Temperature value around the field        :', font=("Helvetica", 20)), sg.Input(font=("Helvetica", 20),size = (20,1)), sg.Text('*C', font=("Helvetica", 20))], 
         [sg.Text('Enter average percentage of Humidity around the field :', font=("Helvetica", 20)), sg.Input(font=("Helvetica", 20),size = (20,1)), sg.Text('%', font=("Helvetica", 20))], 
         [sg.Text('Enter PH value of the soil                                            :', font=("Helvetica", 20)), sg.Input(font=("Helvetica", 20),size = (20,1)), sg.Text('0 To 14', font=("Helvetica", 20))], 
         [sg.Text('Enter average amount of Rainfall around the field        :', font=("Helvetica", 20) ), sg.Input(font=("Helvetica", 20),size = (20,1)),sg.Text('mm', font=("Helvetica", 20))],
         [sg.Text('Enter water-availability                                                :', font=("Helvetica", 20) ), sg.Input(font=("Helvetica", 20),size = (20,1)),sg.Text('lts',font=("Helvetica", 20))],
         [sg.Text(size=(50,1),font=("Helvetica",20) , text_color = 'yellow', key='-OUTPUT1-' )],
         [sg.Button('Submit', font=("Helvetica", 20)),sg.Button('Quit', font=("Helvetica", 20))] ]
window = sg.Window('Crop Recommendation System', layout) 
```

```python
while True: 
    event, values = window.read()
    if event == sg.WINDOW_CLOSED or event == 'Quit':                                                                                            
        break
    print(values[0])
    nitrogen_content =         values[0]                                                                                                        
    phosphorus_content =       values[1]                                                                                                       
    potassium_content =        values[2]                                                                                                        
    temperature_content =      values[3]                                                                                                        
    humidity_content =         values[4]                                                                                                         
    ph_content =               values[5]                                                                                                        
    rainfall =                 values[6]
    water_availability =       values[7]
    
    predict1 = np.array([nitrogen_content,phosphorus_content, potassium_content, temperature_content, humidity_content, ph_content, rainfall, water_availability])
    print(predict1)                                                                                                                            
    predict1 = predict1.reshape(1,-1)                                                                              
    print(predict1)                                                                                                
    predict1 = model.predict(predict1)                                                                              
    print(predict1)                                                                                                
    crop_name = str()
    if predict1 == 0:                                                                                              
        crop_name = 'rice'
    elif predict1 == 1:
        crop_name = 'maize'
    elif predict1 == 2:
        crop_name = 'chickpea'
    elif predict1 == 3:
        crop_name = 'kidneybeans'
    elif predict1 == 4:
        crop_name = 'piegonpeas'
    elif predict1 == 5:
        crop_name = 'mothbeans'
    elif predict1 == 6:
        crop_name = 'mungbean'
    elif predict1 == 7:
        crop_name = 'blackgram'
    elif predict1 == 8:
        crop_name = 'lentil'
    elif predict1 == 9:
        crop_name = 'pomegranate'
    elif predict1 == 10:
        crop_name = 'banana'
    elif predict1 == 11:
        crop_name = 'mango'
    elif predict1 == 12:
        crop_name = 'grapes'
    elif predict1 == 13:
        crop_name = 'watermelon'
    elif predict1 == 14:
        crop_name = 'muskmelon'
    elif predict1 == 15:
        crop_name = 'apple'
    elif predict1 == 16:
        crop_name = 'orange'
    elif predict1 == 17:
        crop_name = 'papaya'
    elif predict1 == 18:
        crop_name = 'coconut'
    elif predict1 == 19:
        crop_name = 'cotton'
    elif predict1 == 20:
        crop_name = 'Sapota'
    elif predict1 == 21:
        crop_name = 'fingermillet'
    

    if int(humidity_content) >=1 and int(humidity_content)<= 33 :                                                
        humidity_level = 'low humid'
    elif int(humidity_content) >=34 and int(humidity_content) <= 66:
        humidity_level = 'medium humid'
    else:
        humidity_level = 'high humid'

    if int(temperature_content) >= 0 and int(temperature_content)<= 6:                                          
        temperature_level = 'cool'
    elif int(temperature_content) >=7 and int(temperature_content) <= 25:
        temperature_level = 'warm'
    else:
        temperature_level= 'hot' 

    if int(rainfall) >=1 and int(rainfall) <= 100:                                                              
        rainfall_level = 'less'
    elif int(rainfall) >= 101 and int(rainfall) <=200:
        rainfall_level = 'moderate'
    elif int(rainfall) >=201:
        rainfall_level = 'heavy rain'
    
    if float(water_availability) >=1 and float(water_availability) <=100:
        water_level = 'less'
    elif float(water_availability) >=101 and float(water_availability) <=1000:
        water_level = 'moderate'
    elif float(water_availability) >=1001:
        water_level = 'abundant'
        
    if int(nitrogen_content) >= 1 and int(nitrogen_content) <= 50:                                            
        nitrogen_level = 'less'
    elif int(nitrogen_content) >=51 and int(nitrogen_content) <=100:
        nitrogen_level = 'not too less but also not too high'
    elif int(nitrogen_content) >=101:
        nitrogen_level = 'high'

    if int(phosphorus_content) >= 1 and int(phosphorus_content) <= 50:                                         
        phosphorus_level = 'less'
    elif int(phosphorus_content) >= 51 and int(phosphorus_content) <=100:
        phosphorus_level = 'not to less but also not to high'
    elif int(phosphorus_content) >=101:
        phosphorus_level = 'high'

    if int(potassium_content) >= 1 and int(potassium_content) <=50:                                          
        potassium_level = 'less'
    elif int(potassium_content) >= 51 and int(potassium_content) <= 100:
        potassium_level = 'not to less but also not to high'
    elif int(potassium_content) >=101:
        potassium_level = 'high'

    if float(ph_content) >=0 and float(ph_content) <=5:                                                        
        phlevel = 'acidic' 
    elif float(ph_content) >= 6 and float(ph_content) <= 8:
        phlevel = 'neutral'
    elif float(ph_content) >= 9 and float(ph_content) <= 14:
        phlevel = 'alkaline'
    
    print(crop_name)
    print(humidity_level)
    print(temperature_level)
    print(rainfall_level)
    print(water_level)
    print(nitrogen_level)
    print(phosphorus_level)
    print(potassium_level)
    print(phlevel)

    
    speak("The ratio of nitrogen in the soil is  " + nitrogen_level + ". The ratio of phosphorus in the soil is  " + phosphorus_level + ". The ratio of potassium in the soil is  " + potassium_level + ". The temperature level around the field is  " + temperature_level + ". The humidity level around the field is  " + humidity_level + ". The ph type of the soil is  " + phlevel + ". The amount of rainfall is  " + rainfall_level + ". The water-availability is  " + water_level )
    window['-OUTPUT1-'].update('The best crop that you can grow : ' + crop_name )                                    
    speak("The best crop that you can grow is  " + crop_name)                                                         


    

window.close()
```
